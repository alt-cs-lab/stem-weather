Start by loading the data into Jupyter. This requires us to read the data from the CSV file, transpose it from columns to rows, and place it in a Pandas dataframe. Let's do that for the Cherokee data!

```python
# Import Libraries
import pandas as pd

# Load the Cherokee CSV file into a DataFrame
cherokee = pd.read_csv('cherokee_cleaned.csv')

# View the first 5 rows
print(cherokee.head())
```

Next, we can graph that data using the Plotly library.

```python
import plotly.express as px

fig = px.line(
  cherokee,
  x='Timestamp', 
  y='SoilTemperature5cmMin', 
)

fig.show()
```

Let's add some helpful labels and markers to the graph. 

```python
fig = px.line(cherokee,
  x='Timestamp', 
  y=['SoilTemperature5cmMin', 'SoilTemperature5cmMax'], 
  title='Soil Temperature at 5cm Depth',
  labels={
    'x': 'Day of Year', 
    'variable': 'Measurement'
    }
  )
fig.update_xaxes(title_text="Day of Year")
fig.update_yaxes(title_text='Soil Temperature (°F)')
fig.data[0].name = "5cm Min"
fig.data[1].name = "5cm Max"
fig.show()
```

Now we can add some trendlines and fix the days of the year to match the months

```python
fig = px.line(cherokee,
  x='Timestamp', 
  y=['SoilTemperature5cmMin', 'SoilTemperature5cmMax'], 
  title='Soil Temperature at 5cm Depth',
  labels={
    'x': 'Day of Year', 
    'variable': 'Measurement'
    }
  )
fig.update_xaxes(title_text="Day of Year")
fig.update_yaxes(title_text='Soil Temperature (°F)')
fig.data[0].name = "5cm Min"
fig.data[1].name = "5cm Max"

fig.add_hline(
  y=50, 
  line_dash="dash", 
  line_color="red", 
  annotation_text="Corn", 
  annotation_position="top left"
)
fig.add_hline(
  y=55, 
  line_dash="dash", 
  line_color="blue", 
  annotation_text="Soybeans",
  annotation_position="top left"
)

fig.update_xaxes(
  tickmode='array', 
  tickvals=[0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334], 
  ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)

fig.show()
```

Now let's load the rest of the data by copy/pasting what we did above!

```python
# Load the other three CSV files into DataFrames
cheyenne = pd.read_csv('cheyenne_cleaned.csv')
hiawatha = pd.read_csv('hiawatha_cleaned.csv')
richfield = pd.read_csv('richfield_cleaned.csv')
```

Now we can create a combined data frame for all of this data!

```python
cherokee['Location'] = 'Cherokee'
cheyenne['Location'] = 'Cheyenne'
hiawatha['Location'] = 'Hiawatha'
richfield['Location'] = 'Richfield'
combined_df = pd.concat([cherokee, cheyenne, hiawatha, richfield], ignore_index=True)
```

And now we can graph it!

```python
fig=px.line(combined_df,
  x='Timestamp', 
  y='SoilTemperature5cmMin', 
  color='Location',
  title='Min Soil Temperature at 5cm Depth',
  labels={'x': 'Day of Year', 'y': 'Soil Temperature (°F)', 'variable': 'Measurement'},
  )
fig.show()
```

Adding in trendlines and better titles.

```python
fig=px.line(combined_df,
  x='Timestamp', 
  y='SoilTemperature5cmMin', 
  color='Location',
  title='Min Soil Temperature at 5cm Depth',
  labels={'x': 'Day of Year', 'y': 'Soil Temperature (°F)', 'variable': 'Measurement'},
  )
fig.add_hline(
  y=50, 
  line_dash="dash", 
  line_color="red", 
  annotation_text="Corn", 
  annotation_position="top left"
)
fig.add_hline(
  y=55, 
  line_dash="dash", 
  line_color="blue", 
  annotation_text="Soybeans",
  annotation_position="top left"
)
fig.update_xaxes(
  tickmode='array', 
  tickvals=[0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334], 
  ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
fig.show()
```
